## Task 2.2
$$
\newcommand{\del}[2]{\frac{\partial #1}{\partial #2}}  % Partial derivative
\newcommand{\ddel}[2]{\frac{\partial^2 #1}{\partial #2^2}}  % Partial second derivative
\newcommand{\deriv}[2]{\frac{\mathrm{d} #1}{\mathrm{d} #2}}  % Derivative
\newcommand{\dderiv}[2]{\frac{\mathrm{d}^2 #1}{\mathrm{d} #2^2}}  % Second derivative
$$
Let $t = t' t_0$ and $x=x' x_0$.
Setting $x_0 = L$ we get
$$
-\frac{\hbar^2}{2mL^2} \ddel{\Psi}{x'} = \frac{i\hbar}{t_0} \del{\Psi}{t'},
$$
and we see that setting $t_0 = \frac{2mL^2}{\hbar}$ gives us
$$
-\ddel{\Psi}{x'} = i \del{\Psi}{t'}.
$$


## Task 2.3
(2.7) therefor gives
$$
-\ddel{\psi_n}{x'} = \frac{E_n 2mL^2}{\hbar^2} \psi_n
$$
or
$$
\ddel{\psi_n}{x'} = \lambda_n \psi_n,
$$
where we defined
$$
\lambda_n = \frac{2mL^2}{\hbar^2} E_n.
$$

```julia
# 2.4

# (2.9)
# -ddel_x' psi_n = lambda_n psi_n
# Use central finite difference method
# ddel_x f(x) ~= (-f(x-h) + 2f(x) - f(x+h))/h^2
# The boundary conditions are psi = 0, so the system to solve has N-2 points

function solve_eigenfunction_value(N)
    x = range(0, 1, length=N)
    dx = x.step.hi

    # Allocate array for vectors, solve and scale, and set end and start to zero
    vectors = Array{Float64, 2}(undef, N, N-2)
    H = SymTridiagonal(fill(2, N-2), fill(-1, N-3))
    values, vectors[2:end-1, :] = eigen(H)
    values ./= dx^2
    vectors ./= sqrt(dx)
    vectors[[1,end], :] .= 0
    for i in 1:N-2
        vectors[:, i] = flipsign.(vectors[:, i], vectors[2, i])  # Ensure we get +sin, not -sin
    end
    
    return values, vectors, dx
end

values, vectors, dx = solve_eigenfunction_value(500)
```

```julia
# 2.6

function eigenfunction_error(vector::AbstractVector, n::Integer)
    """Finds the sum of squares error in vector"""
    x = range(0, 1, length=length(vector))
    anal = sqrt(2)*sin.(n*pi*x)
    
    # Return the "mean" of square error
    #return sum(abs2, vector-anal)/length(vector)
    error = (vector-anal)./max.(anal, 0.001)
    return mean(abs2, error)
end


function plot_functions(values, vectors, ns; scale=30, analytical=true)
    # psi_n(x') = sqrt(2) sin(n pi x')
    N, num = size(vectors)
    x = range(0, 1, length=N)
    for (i,n) in enumerate(ns)
        plt.axhline(values[i], linestyle="dashed", color="gray")
        if analytical
            plt.plot(x, (pi*n)^2 .+ scale * sqrt(2)*sin.(n*pi*x), label="Analytical \$\\psi_n(x')\$")
        end
        plt.plot(x, values[i] .+ scale * vectors[:, i], label="Numerical")
    end
    plt.legend()
end


to_check = [1, 5, 8, 10, 12]  # List of functions to check
plot_functions(values[to_check], vectors[:, to_check], to_check, scale=50)
```

## Task 2.7

$$
-\frac{i E_n t}{\hbar} =
-\frac{i \lambda_n \hbar t'}{\hbar} = 
-i\lambda t',
$$
where the definition of $t'$ and $\lambda_n$ was used.

```julia
# 2.8
function coefficient(Psi_0, psi_n, dx)
    # Assume psi_n real
    return sum(Psi_0.*psi_n)*dx
end

function coefficient2(Psi_0, psi_n, dx)
    """Uses Trapezoidal rule"""
    sum = 0.0
    for i in eachindex(Psi_0)[2:end-1]
        sum += Psi_0[i]*psi_n[i]
    end
    sum += 0.5 * (Psi_0[1]*psi_n[1] + Psi_0[end]*psi_n[1])
    sum *= dx
    return sum
end
```

```julia
# 2.10

function develop_in_time(t, vectors, coefficients, values)
    total = zero(vectors[:, 1])
    for n in eachindex(coefficients)
        total += coefficients[n] * vectors[:, n] * exp(values[n]*t*im)
    end
    return total
end


# Psi_0 = sqrt(2)sin(pi x) (= psi_1)
N = 200
num_eigenfunctions = 20 

values, vectors, dx = solve_eigenfunction_value(N)
Psi_0 = sqrt(2) * sin.(pi*x)
coefficients = [coefficient(Psi_0, vectors[:, n], dx) for n in 1:num_eigenfunctions]

t_list = 0:0.02:4
v = Array{Complex, 2}(undef, N, length(t_list)+1)
v[:, 1] = Psi_0
for i_t in eachindex(t_list)
    v[:, i_t+1] = develop_in_time(t_list[i_t], vectors[:, 1:num_eigenfunctions], coefficients, values[1:num_eigenfunctions])
end
```

```julia
# 3.1

function solve_eigenfunction_value(N, V0=0)
    x = range(0, 1, length=N)
    dx = x.step.hi

    # Allocate array for vectors, solve and scale, and set end and start to zero
    vectors = Array{Float64, 2}(undef, N, N-2)
    diagonal = fill(2.0/dx^2, N-2)
    subdiagonal = fill(-1.0/dx^2, N-3)
    idx = eachindex(diagonal)/length(diagonal)
    diagonal[1/3 .<= idx .<= 2/3] .+= V0
    
    H = SymTridiagonal(diagonal, subdiagonal)
    values, vectors[2:end-1, :] = eigen(H)
    vectors ./= sqrt(dx)
    vectors[[1,end], :] .= 0
    for i in 1:N-2
        vectors[:, i] = flipsign.(vectors[:, i], vectors[2, i])  # Ensure we get +sin, not -sin
    end
    
    return values, vectors, dx
end
```

```julia
# 3.2

function solve_eigenfunction_value2(N_third, V0=0)    
    N = 3*N_third
    dx = 1/(N-1)
    
    # Don't inclue endpionts because of BC at 0 and 1
    diagonal = fill(2.0/dx^2, N-2)
    diagonal[N_third:2*N_third-1] .+= V0
    subdiagonal = fill(-1.0/dx^2, N-3)
    
    vectors = Array{Float64, 2}(undef, N, N-2)
    H = SymTridiagonal(diagonal, subdiagonal)
    values, vectors[2:end-1, :] = eigen(H)
    vectors ./= sqrt(dx)  # Normalize
    vectors[[1,end], :] .= 0
    for i in 1:N-2
        vectors[:, i] = flipsign.(vectors[:, i], vectors[2, i])  # Ensure we get +sin, not -sin
    end
    
    return values, vectors, dx
end


N_third = 300  # Points for one third, ie one part linear part of the potential
N = 3*N_third
mode_range = 1:4
v0 = 1e3
values, vectors, dx = solve_eigenfunction_value2(N_third, v0)
plt.plot([0,1/3,1/3,2/3,2/3,1], [0,0,v0,v0,0,0], color="gray", linestyle="dashed")
plot_functions(values[mode_range], vectors[:, mode_range], mode_range, analytical=false)
```

```julia
# 3.4
function eigenvalues_at_zero(lambda, v0)
    k = sqrt(lambda)
    kappa = sqrt(v0 - lambda)
    first_term = exp(kappa/3) * (kappa*sin(k/3) + k*cos(k/3))^2
    second_term = exp(-kappa/3) * (kappa*sin(k/3) - k*cos(k/3))^2
    return first_term - second_term
end


function regula_falsi(f, epsilon, a, b)
    fa = f(a)
    fb = f(b)
    c = (a*fb - b*fa)/(fb - fa)
    fc = f(c)
    iterations = 0
    while abs(fc) > epsilon
        iterations += 1
        if sign(fc*fb) == -1
            a=c
            fa = fc
        else
            b=c
            fb = fc
        end
        c = (a*fb - b*fa)/(fb - fa)
        fc = f(c)
    end
    return c, iterations
end
```

```julia
# 4.1

# Time dependent potential
#           { 0 for 0 < x < L/3
#           { V0 for L/3 < x < 2L/3
# v(x, t) = { Vr(t) for 2L/3 < x < L
#           { infty otherwise

function get_detuned_hamil(vr, num_points_per_well=40; v0=100.0)
    N = num_points_per_well * 3
    dx = 1.0 / (N-1)
    second_well = num_points_per_well+1
    third_well = 2*num_points_per_well+1

    V = zeros(N)
    V[second_well:third_well-1] .= v0
    V[third_well:end] .= vr

    # Hamiltonian: -d^2/dx^2 + v(x) approximated with central finite difference
    diagonal = fill(2.0/dx^2, N)
    diagonal .+= V
    subdiagonal = fill(-1.0/dx^2, N-1)
    H = SymTridiagonal(diagonal, subdiagonal)
end

function get_detuned_energies(vr, num_points_per_well=40)
    H = get_detuned_hamil(vr, num_points_per_well)
    # Solve the system
    return eigvals(H, 1:2)
end

function get_detuned_states(vr, num_points_per_well=40)
    H = get_detuned_hamil(vr, num_points_per_well)
    energies = eigvals(H, 1:2)
    states = eigvecs(H, energies)
    ground_state = states[:, 1]
    excited_state = states[:, 2]
    return ground_state, excited_state
end

vr_range = -10:1:10
energies = Array{Float64, 2}(undef, 2, length(vr_range))
for (i, vr) in enumerate(vr_range)
    energies[:, i] = get_detuned_energies(vr, 100)
end
```

# Solving the Volterra integral

```julia
# Get the two first states of the unperturbed system
ground, excited = get_detuned_states(0)

omega = 1
epsilon = omega
tau = 0.02*epsilon

dt = 0.1
end_time = 1000
t_space = 0:dt:end_time

psi_0 = [1; 0]  # Start in the ground state, and evolve in time
psi = Array{Complex, 2}(undef, 2, length(t_space))
psi[:, 1] = psi_0

# Call signature of HI: HI(t, tau=1, eps=1, omega=1)

function crank_nicolson_step(state, t, dt, tau, eps, omega)
    RHS = (I - im*dt/2 * HI(t, tau=tau, eps=eps, omega=omega)) * state
    new_state = (I + im*dt/2 * HI(t+dt, tau=tau, eps=eps, omega=omega)) \ RHS
    return new_state
end

for i in eachindex(t_space)[2:end]
    psi[:, i] = crank_nicolson_step(psi[:, i-1], t_space[i-1], dt, tau, epsilon, omega)
end
```

```julia
state = Array{Complex, 2}(undef, length(ground), length(t_space))
for i in eachindex(t_space)
    state[:, i] = psi[1, i]*ground + psi[2, i]*excited
end
```
